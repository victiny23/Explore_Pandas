```python
import pandas as pd
```

```python
# read from csv file
df = pd.read_csv('C:/Users/Victiny/Python_Project/Explore_Pandas/data_example/wine.csv')
```

```python
# This is similar to df['points'].value_counts()
df.groupby('points').points.count()
df.groupby('points')['points'].count()
```

```python
# or use size()
df.groupby('points').size()
```

```python
df.groupby('points').points.min()
df.groupby('points')['points'].min()
```

```python
# get the name of the first wine reviewed from each winery
df.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# you may group by multiple columns
# pick out the best wine by country and province
# group by country and province
# use idxmax to find the index of the row with the highest points in each group and return that row from each group
df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# agg() lets you run a bunch of different functions on your DataFrame simultaneously
df.groupby('country').price.agg([len, min, max, sum])
df.groupby('country')['price'].agg([len, min, max, sum, 'mean'])
```

```python
df1 = pd.DataFrame({'A': [1, 1, 2, 2],
                    'B': [1, 2, 3, 4],
                    'C': [7, 8, 9, 12]})
df1
```

```python
# agg() for multiple columns 
df1.groupby('A').agg(['min', 'max', 'sum', 'mean'])
```

```python
# agg() for a single column
df1.groupby('A')['B'].agg(['min', 'max', 'sum', 'mean'])
```

```python
# different aggregation per column
df1.groupby('A').agg({'B': ['min', 'max'], 'C': ['idxmax', 'idxmin', 'median']})
```

```python
# aggregate over rows (default)
df1.agg('mean')
df1.agg('mean', axis=0)
df1.agg('mean', axis='rows')
```

```python
# aggregate over columns
df1.agg('mean', axis='columns')
df1.agg('mean', axis=1)
```

```python
"""Multi-indexes"""
countries_reviewed = df.groupby(['country', 'province']).agg({'price': 'mean', 'points': 'mean'})
print(type(countries_reviewed.index))
countries_reviewed
```

```python
countries_reviewed.reset_index()
```

```python
"""Sorting"""
# sort_values() defaults to an ascending sort - lowest values on top
countries_reviewed.sort_values(by='price')
countries_reviewed.sort_values(by='price', ascending=False)
```

```python
# sorting on multiple columns
countries_reviewed.sort_values(by=['price', 'points'], ascending=[False, False])
```

```python
# what combination of countries and varieties are most common?
# pick any column and aggreagate by length, then sort by values
country_variety_counts = df.groupby(['country', 'variety']).title.agg(len).sort_values(ascending=False)
# or use size()
country_variety_counts = df.groupby(['country', 'variety']).size().sort_values(ascending=False)
country_variety_counts
```

```python
# sorting by index values (string)
countries_reviewed.sort_index()
```

```python
# sorting by index values (number)
# find the row with highest points for each price
df.groupby('price').points.agg('max').sort_index
```
