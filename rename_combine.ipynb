```python
"""Combining"""
# concat() given a list of elements, this function will smush those elements together along an axis
canadian_youtube = pd.read_csv('c:/Users/Victiny/Python_Project/Explore_Pandas/data_example/ca_youtube.csv', index_col=0)
british_youtube = pd.read_csv('c:/Users/Victiny/Python_Project/Explore_Pandas/data_example/gb_youtube.csv', index_col=0)
pd.concat([canadian_youtube, british_youtube]).head()
```

```python
canadian_youtube[canadian_youtube['title'].isnull()]['title'].dtype
```

```python
# join() lets you combine different DataFrame objects which have an index in common
canadian_youtube['title'].fillna('Unknown').astype('str')
british_youtube['title'].fillna('Unknown').astype('str')
canadian_youtube['trending_date'].fillna('Unknown').astype('str')
british_youtube['trending_date'].fillna('Unknown').astype('str')

left1 = canadian_youtube.set_index(['title', 'trending_date'])
right1 = canadian_youtube.set_index(['title', 'trending_date'])
left1.head(3)
```

```python
right1.index.get_level_values(1)
```

```python
left1.join(right1, lsuffix='_CA', rsuffix='_GB')
```
