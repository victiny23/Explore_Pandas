```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'name': ['Jack', 'Jack', 'Jack', 'Tom', 'Tom', 'Tom', 'David', 'David', 'David'],
                   'subject': ['Python', 'C', 'SQL']*3,
                   'score': [95, 60, 95, 96, 95, 80, 99, 94, 88]
                   })
df1
```

```python
""" Row to Column"""
df1.pivot(index='name', columns='subject', values='score')
```

```python
""" Handling repeated values"""
df2 = pd.concat([pd.DataFrame({'name': ['Jack'], 'subject': ['Python'], 'score':[ 95]}, columns=['name', 'subject', 'score']), df1])
df2
```

```python
# ValueError: Index contains duplicate entries, cannot reshape
df2.pivot(index='name', columns='subject', values='score')
```

```python
# remove the duplicates before pivotting
df2.drop_duplicates().pivot(index='name', columns='subject', values='score')
```

```python
# pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean')
pd.pivot_table(df2, index='name', columns='subject', values='score')
```

```python
pd.pivot_table(df2, index='name', columns='subject', values='score', aggfunc={'score': 'max'})
```

```python
# calculate the total score for everyone
df_pivot1 = pd.pivot_table(df2, index='name', columns='subject', values='score', aggfunc={'score': 'max'})
df_pivot1['total'] = df_pivot1.apply(lambda x: np.sum(x), axis=1)
df_pivot1
```

```python
df_pivot2 = pd.pivot_table(df2, index='name', columns='subject', values='score', aggfunc={'score': 'max'})
df_pivot2['total'] = df_pivot2.sum(axis=1)
df_pivot2
```

```python
total = pd.pivot_table(df1, index='name', values='score', aggfunc='sum')
total
```

```python
pd.pivot_table(df1, index='name', columns='subject', values='score').join(total)
```

```python
total1 = pd.pivot_table(df1, index='name', values='score', aggfunc='sum').rename({'score':'total score'}, axis=1)
pd.pivot_table(df1, index='name', columns='subject', values='score').join(total1)
```
